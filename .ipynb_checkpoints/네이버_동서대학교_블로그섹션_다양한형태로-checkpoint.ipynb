```python
from bs4 import BeautifulSoup     
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import sys
import pandas as pd
import xlwt

query_txt = input('크롤링할 키워드는 무엇입니까?: ')

ft_name = input('검색 결과를 저장할  txt 파일경로와 이름을 지정하세요(예:c:\\data\\test.txt): ')
fc_name = input('검색 결과를 저장할  csv 파일경로와 이름을 지정하세요(예:c:\\data\\test.csv): ')
fx_name = input('검색 결과를 저장할  xls 파일경로와 이름을 지정하세요(예:c:\\data\\test.xls): ')

path = "C:\Temp\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://www.naver.com")
time.sleep(2)

element = driver.find_element(By.ID,"query")
element.send_keys(query_txt)
driver.find_element(By.XPATH, '//*[@id="sform"]/fieldset/button').click()

time.sleep(4)

driver.find_element(By.LINK_TEXT,"VIEW").click()

time.sleep(2)
driver.find_element(By.LINK_TEXT,"블로그").click()

full_html = driver.page_source

soup = BeautifulSoup(full_html, 'html.parser')

content_list = soup.find('ul',class_='lst_total')

orig_stdout = sys.stdout
f = open(ft_name , 'a' , encoding='UTF-8')
sys.stdout = f
time.sleep(1)

no = 1
no2 = []
title2 = []
date2 = []
writer2 = []
content2 = []

for i in content_list.find_all('li', "bx"):
    if no < 11:
        no2.append(no)
        print('번호:',no)

        title = i.find('a', 'api_txt_lines total_tit').get_text()
        title2.append(title)
        print('제목:', title.strip())

        date = i.find('span', 'sub_time sub_txt').get_text()
        date2.append(date)
        print('작성일자:', date.strip())

        writer = i.find('a', 'sub_txt sub_name').get_text()
        writer2.append(writer)
        print('작성자:', writer.strip())
        
        content = i.find('div', 'api_txt_lines dsc_txt').get_text()
        content2.append(content)
        print('내용', content.strip())
        
        print("\n")

        no += 1
    else:
        break

sys.stdout = orig_stdout
f.close()

df = pd.DataFrame()

df['번호'] = no2
df['제목'] = title2
df['작성일자'] = date2
df['작성자'] = writer2
df['내용'] = content2

df.to_csv(fc_name, encoding="utf-8-sig")
df.to_excel(fx_name)

print(" 요청하신 데이터 수집 작업이 정상적으로 완료되었습니다")
```
