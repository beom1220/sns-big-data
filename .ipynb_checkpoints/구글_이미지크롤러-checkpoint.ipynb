```python
from bs4 import BeautifulSoup     
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import sys
import pandas as pd
import xlwt
import os
import urllib.request
import urllib

print("============================================================================================================")
print("Google 사이트에서 이미지를 검색하여 수집하는 크롤러입니다.")
print("============================================================================================================")
keyword = input('1. 크롤링할 키워드는 무엇입니까?(예:여행): ')
num = int(input('2. 크롤링할 건수는 몇 건입니까?: '))


f_dir = input("3.결과 파일을 저장할 폴더명만 쓰세요(예:c:\\data\\):")

n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)

os.makedirs(f_dir+s+'-'+keyword)
os.chdir(f_dir+s+'-'+keyword)

path = "C:\Temp\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://www.google.com/imghp")

time.sleep(2)

search_box = driver.find_element(By.CLASS_NAME, 'gLFyf')
search_box.send_keys(keyword)
search_box.submit()
time.sleep(2)

count = 0

while count < num:
    time.sleep(2)
    
    for _ in range(int(num/20)):
        body = driver.find_element(By.CSS_SELECTOR, 'body')
        body.send_keys(Keys.END)
        time.sleep(2)
    
    images = driver.find_elements(By.CLASS_NAME, 'rg_i')
    
    for img in images:
        time.sleep(2)
        try:
            img_url = img.get_attribute('src')
            file_name = str(count) + ".jpg"
            urllib.request.urlretrieve(img_url, file_name)
            print(count, "번째 이미지 다운로드 완료.")
            count += 1
            
            if count >= num:
                break
        except:
            print("이미지 로딩에 실패했습니다.")
        
driver.quit()
print("이미지 다운로드가 완료되었습니다.")
```
